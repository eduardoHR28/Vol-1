```python
"""
  --Two Digits Calculator--
    Receive two entries, show the calculator interface and, based in the users's
  choice, execute the proper operation, allowing the user to repeat or exit.

"""
import random
#Defining the calculator mathematical operations
def sum(a,b):
  return a+b
def subtraction(a,b):
  return a-b
def multiplication(a,b):
  return a*b
def divisision(a,b):
  return a/b
def exponetiation(a,b):
  return a**b
def chosen(a,b):
  return random.choice([a,b])

#Print the head of the calculator
print('||'*9,'CALCULATOR 2000','||'*9)

#Ask to user for two numbers
a = float(input('First digit: '))
b = float(input('Second digit: '))

#Defines a dictionary empty to save future operations
dic={}

#Loop the calculator interface until the user decline to proceed
while True:
  #Show to user what operations it can do
  print(
      '\nOPERATIONS:\n'
      '[1]-SUM\n'
      '[2]-SUBTRACTION\n'
      '[3]-MULTIPLICATION\n'
      '[4]-DIVISION\n'
      '[5]-EXPONENTIATION\n'
      '[6]-CHOSEN'
  )
  o = int(input('OPERATION NUMBER: '))

#Conditionally to input, execute one of the function:
  if o == 1:
    dic['SUM'] = sum(a,b)
    print('SUM: ', sum(a,b))    #Add the sum with the consecutive key

  if o == 2:
    print('SUBTRACTION: ', subtraction(a,b))
    dic['SUBTRACTION'] = subtraction(a,b)

  if o == 3:
    print('MULTIPLICATION: ', multiplication(a,b))
    dic['MULTIPLICATION'] = multiplication(a,b)

  if o == 4:
    if b == 0:
      print('ERROR: NO DIVISION BY 0.')       #The division by zero solution
    else:
      print('DIVISION: ', divisision(a,b))
    dic['DIVISION'] = divisision(a,b)

  if o == 5:
    print('EXPONENTIATION: ', exponetiation(a,b))
    dic['EXPONENTIATION'] = exponetiation(a,b)

  if o == 6:
    print('MY CHOSEN ONE: ', chosen(a,b))   #Randomly choose between the inputs
    dic['CHOSEN'] = chosen(a,b)

#The user can exit answering "n", breaking the While loop
  if (input('\nContinue(s/n)?')) == 'n':
    print(dic.items())
    print(' '*8,'!!Thank you for using our services!!')
    print('||'*12, 'END','||'*12)
    break
```
